```python
category_sales=(df.groupby("Category")["Sales"].sum().sort_values(ascending=False))
```

```python
category_sales
```

```python
category_profit=(df.groupby("Category")["Profit"].sum().sort_values(ascending=False))
```

```python
category_profit
```

## Category Performance Analysis

### Finding
Technology generated the highest sales (USD 836K) and profit (USD 145K). Furniture generated strong sales (USD 742K) but contributed only $18K in profit.

### Insight
Furniture is a high-revenue but low-profit category, indicating potential pricing, discounting, or cost-related issues. Technology and Office Supplies contribute significantly more profit relative to their sales.

### Recommendation
Review pricing and discount strategies for Furniture products while increasing focus on high-margin Technology and Office Supplies categories.

```python
df["Order Date"]=pd.to_datetime(df["Order Date"])
```

```python
df.head()
```

```python
df["YearMonth"]=df["Order Date"].dt.to_period("M")
```

```python
monthly_sales = df.groupby("YearMonth")["Sales"].sum()
```

```python
monthly_sales
```

```python
monthly_sales.plot(figsize=(12,5))
plt.title("Monthly sales trend")
plt.ylabel("sales")
plt.grid(alpha=0.3)
plt.show()
```

## Sales Trend Analysis

### Finding
Monthly sales show an overall upward trend from 2014 to 2017, reaching the highest level near the end of 2017.

### Insight
The business demonstrates sustained revenue growth over time, while recurring spikes suggest seasonal demand patterns and high-performing sales periods.

### Recommendation
Analyze the drivers behind peak sales months and replicate successful promotional, inventory, or marketing strategies during future periods.
